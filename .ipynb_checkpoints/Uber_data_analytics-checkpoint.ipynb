```python
import io
import pandas as pd
```

```python
df = pd.read_csv("uber_data.csv")
```

```python
df.head()
#Give top 5 entries
```

```python
df.info()
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

#Changes data type from object to datetime so that we can excess data like date, hours etc
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)

#It will drop the same columns if any and then reset their index

df['trip_id'] = df.index

#Nmae index as trip_id
```

```python
df.head()
```

```python
datetime_dim = df[['tpep_pickup_datetime','tpep_dropoff_datetime']].reset_index(drop=True)

#Taking two columns 'tpep_pickup_datetime','tpep_dropoff_datetime' to extract data like hour , day etc 
```

```python
datetime_dim['tpep_pickup_datetime'] = datetime_dim['tpep_pickup_datetime']
datetime_dim['pick_hour'] = datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['pick_day'] = datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['pick_month'] = datetime_dim['tpep_pickup_datetime'].dt.month
datetime_dim['pick_year'] = datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['pick_weekday'] = datetime_dim['tpep_pickup_datetime'].dt.weekday

#Seperated hour, day, month, year, weekday from 'tpep_pickup_datetime' and named them 
```

```python
datetime_dim['tpep_dropoff_datetime'] = datetime_dim['tpep_dropoff_datetime']
datetime_dim['drop_hour'] = datetime_dim['tpep_dropoff_datetime'].dt.hour
datetime_dim['drop_day'] = datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['drop_month'] = datetime_dim['tpep_dropoff_datetime'].dt.month
datetime_dim['drop_year'] = datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['drop_weekday'] = datetime_dim['tpep_dropoff_datetime'].dt.weekday

#Seperated hour, day, month, year, weekday from 'tpep_dropoff_datetime' and named them 
```

```python
datetime_dim['datetime_id'] = datetime_dim.index

#A new index is assigned and named after datetime_id
```

```python
# datetime_dim = datetime_dim.rename(columns={'tpep_pickup_datetime': 'datetime_id'}).reset_index(drop=True)
datetime_dim = datetime_dim[['datetime_id', 'tpep_pickup_datetime', 'pick_hour', 'pick_day', 'pick_month', 'pick_year', 'pick_weekday',
                             'tpep_dropoff_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
# Reordering of columns 
datetime_dim.head()
```

```python
passenger_count_dim = df[['passenger_count']].reset_index(drop=True)
#adds 'passenger_count' to passenger_count_dim dataframe
passenger_count_dim['passenger_count_id'] = passenger_count_dim.index
#assigns index to passenger_count_id
passenger_count_dim = passenger_count_dim[['passenger_count_id','passenger_count']]
#Making a dataframe passenger_count_dim wich includes two columns 'passenger_count_id','passenger_count' also ordering them as shown
passenger_count_dim.head()
```

```python
trip_distance_dim = df[['trip_distance']].reset_index(drop=True)
#adds 'trip_distance' to trip_distance_dim dataframe
trip_distance_dim['trip_distance_id'] = trip_distance_dim.index
#assigns index to trip_distance_id
trip_distance_dim = trip_distance_dim[['trip_distance_id','trip_distance']]
#Making a dataframe trip_distance_dim wich includes two columns 'trip_distance_id','trip_distance' also ordering them as shown
trip_distance_dim.head()
```

```python
rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

#Data dictionary which is available officially in the Uber data sheet

rate_code_dim = df[['RatecodeID']].reset_index(drop=True)
#Adding 'RatecodeID' to the new data frame rate_code_dim and also resetting index 

rate_code_dim['rate_code_id'] = rate_code_dim.index
#Making new column 'rate_code_id' to the rate_code_dim and assigning its index

rate_code_dim['rate_code_name'] = rate_code_dim['RatecodeID'].map(rate_code_type)
#New column rate_code_name which is mapping of RatecodeID to rate_code_type

rate_code_dim = rate_code_dim[['rate_code_id','RatecodeID','rate_code_name']]
#Reordering of all 3 columns

rate_code_dim.head()
```

```python
pickup_location_dim = df[['pickup_longitude', 'pickup_latitude']].reset_index(drop=True)
#Adding 'pickup_longitude', 'pickup_latitude' from df to new dataframe pickup_location_dim and resetting index and dropping all duplicates
pickup_location_dim['pickup_location_id'] = pickup_location_dim.index
#Adding 'pickup_location_id' to the pickup_location_dim dataframe and assigning index
pickup_location_dim = pickup_location_dim[['pickup_location_id','pickup_latitude','pickup_longitude']] 
#Reordering of all 3 columns

pickup_location_dim.head()
```

```python
dropoff_location_dim = df[['dropoff_longitude', 'dropoff_latitude']].reset_index(drop=True)
#Adding 'dropoff_longitude', 'dropoff_latitude' from df to new dataframe dropoff_location_dim and resetting index and dropping all duplicates
dropoff_location_dim['dropoff_location_id'] = dropoff_location_dim.index
#Adding 'dropoff_location_id' to the dropoff_location_dim dataframe and assigning index
dropoff_location_dim = dropoff_location_dim[['dropoff_location_id','dropoff_latitude','dropoff_longitude']]
#Reordering of all 3 columns

dropoff_location_dim.head()
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
#Data dictionary which is available officially in the Uber data sheet

payment_type_dim = df[['payment_type']].reset_index(drop=True)
#Adding 'RatecodeID' to the new data frame rate_code_dim and also resetting index and droppping duplicates

payment_type_dim['payment_type_id'] = payment_type_dim.index
#Making new column 'payment_type_id' to the payment_type_dim dataframe and assigning index

payment_type_dim['payment_type_name'] = payment_type_dim['payment_type'].map(payment_type_name)
#New column payment_type_namee which is mapping of payment_type to payment_type_name

payment_type_dim = payment_type_dim[['payment_type_id','payment_type','payment_type_name']]
#Merging all 3 columns

payment_type_dim.head()
```

```python
fact_table = df.merge(passenger_count_dim, left_on='trip_id', right_on='passenger_count_id') \
             .merge(trip_distance_dim, left_on='trip_id', right_on='trip_distance_id') \
             .merge(rate_code_dim, left_on='trip_id', right_on='rate_code_id') \
             .merge(pickup_location_dim, left_on='trip_id', right_on='pickup_location_id') \
             .merge(dropoff_location_dim, left_on='trip_id', right_on='dropoff_location_id')\
             .merge(datetime_dim, left_on='trip_id', right_on='datetime_id') \
             .merge(payment_type_dim, left_on='trip_id', right_on='payment_type_id') \
             [['trip_id','VendorID', 'datetime_id', 'passenger_count_id',
               'trip_distance_id', 'rate_code_id', 'store_and_fwd_flag', 'pickup_location_id', 'dropoff_location_id',
               'payment_type_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount',
               'improvement_surcharge', 'total_amount']]

#Merging all columns based on their common trait and according to their id which was mapped and made on Lucid
```
